#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
fig, ax = plt.subplots(1,2, figsize=(16,5))

ax[0].plot(x, x**2)
ax[0].plot(x, np.exp(x))
ax[0].legend(labels=['y = x**2', 'y = exp(x)'])
ax[0].set_title('Linear scale', fontweight='bold')

ax[1].plot(x, x**2)
ax[1].plot(x, np.exp(x))
ax[1].legend(labels=['y = x**2', 'y = exp(x)'])
ax[1].set_yscale('log')
ax[1].set_title('Logarithmic scale', fontweight='bold')
```

## Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here
vehicles_df = pd.read_csv('vehicles.csv')
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,6))

ax.scatter(vehicles_df['Combined MPG'], vehicles_df['Highway MPG'])
ax.set_title('Scatter: combined/highway MPG', fontsize=18, fontweight='bold', c='tab:blue')
ax.set_xlabel('Combined MPG', fontweight='bold')
ax.set_ylabel('Highway MPG', fontweight='bold')
```

## Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
# your code here
fig, ax = plt.subplots(1,1, figsize=(3,9))

ax.boxplot(vehicles_df['CO2 Emission Grams/Mile'])
ax.set_title('Box Whisker: CO2 Emission Grams/Mile', fontsize=10, fontweight='bold', c='tab:orange')
ax.set_xticklabels(['CO2 Emission Grams/Mile'], fontsize=10, fontweight='bold')
ax.set_ylabel('Nº of Cars', fontweight='bold')
```

## Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here
fig, ax = plt.subplots(1,1, figsize=(12,6))

ax.hist(vehicles_df['Fuel Barrels/Year'], color='darkslategray')
ax.set_title('Histogram: Fuel Barrels/Year', fontsize=14, fontweight='bold', c='darkslategray')
ax.set_xlabel('Fuel Barrels/Year', fontweight='bold', fontsize=12)
ax.set_ylabel('Nº of Cars', fontweight='bold', fontsize=12)
```

## Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
def ftype_cmpg(types):
    return {typ: vehicles_df[vehicles_df['Fuel Type'] == typ]['City MPG'].mean() for typ in types}
```

```python
f_types = vehicles_df['Fuel Type'].unique()

f_types_mCmpg = ftype_cmpg(f_types)
```

```python
f_types_mCmpg_df = pd.DataFrame.from_dict(f_types_mCmpg, orient='index', columns=['Mean_city_MPG'])
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,6))

ax.bar(f_types_mCmpg_df.index, height=f_types_mCmpg_df['Mean_city_MPG'], color='darkolivegreen')
plt.xticks(rotation=70)
ax.set_title('Mean City MPG per fuel types', fontsize=14, fontweight='bold', c='darkolivegreen')
ax.set_xlabel('Fuel Types', fontweight='bold', fontsize=12)
ax.set_ylabel('Mean City MPG', fontweight='bold', fontsize=12)
```
